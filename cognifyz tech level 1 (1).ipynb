# **TASK 3:  Price Range vs. Online Delivery and Table Booking**

```python
#important all libares which need to be import -
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Load the dataset -
data = pd.read_csv("/content/Dataset .csv")
```

A. Analyze if there is a relationship between the price range and the availability of online delivery and table booking.

```python
# Create cross-tabulations (contingency tables)
online_delivery_cross_tab = pd.crosstab(data['Price range'], data['Has Online delivery'])
table_booking_cross_tab = pd.crosstab(data['Price range'], data['Has Table booking'])
```

```python
# Visualize the relationships using heatmaps
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

sns.heatmap(online_delivery_cross_tab, annot=True, fmt="d", cmap="YlGnBu", ax=axes[0])
axes[0].set_title('Price Range vs. Online Delivery')

sns.heatmap(table_booking_cross_tab, annot=True, fmt="d", cmap="YlGnBu", ax=axes[1])
axes[1].set_title('Price Range vs. Table Booking')

plt.tight_layout()
plt.show()
```

```python
# Calculate Cramer's V for effect size (optional)
def cramers_v(cross_tab):
  import numpy as np
  chi2 = stats.chi2_contingency(cross_tab)[0]
  n = cross_tab.sum().sum()
  phi2 = chi2 / n
  r, k = cross_tab.shape
  phi2corr = max(0, phi2 - ((k - 1) * (r - 1)) / (n - 1))
  rcorr = r - ((r - 1)**2) / (n - 1)
  kcorr = k - ((k - 1)**2) / (n - 1)
  return np.sqrt(phi2corr / min((kcorr - 1), (rcorr - 1)))

#Import stats from scipy
from scipy import stats

online_delivery_cramers_v = cramers_v(online_delivery_cross_tab)
table_booking_cramers_v = cramers_v(table_booking_cross_tab)

print(f"Cramer's V for Price Range vs. Online Delivery: {online_delivery_cramers_v:.2f}")
print(f"Cramer's V for Price Range vs. Table Booking: {table_booking_cramers_v:.2f}")
```
